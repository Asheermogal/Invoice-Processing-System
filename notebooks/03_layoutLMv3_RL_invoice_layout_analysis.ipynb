# Step 2: Data Loading and Preprocessing

```python
# Step 2: Data Loading and Preprocessing
def load_and_preprocess_data():
    print("Loading and preprocessing data...")
    dataset = load_dataset("nielsr/funsd", trust_remote_code=True)
    processor = LayoutLMv3Processor.from_pretrained("microsoft/layoutlmv3-base", apply_ocr=False)
    print("Data loaded and preprocessed")
    return dataset, processor

def augment_image(image):
    """
    Apply a series of random augmentations to the input image.

    :param image: PIL Image object
    :return: Augmented PIL Image object
    """
    # Convert to RGB if it's not
    if image.mode != 'RGB':
        image = image.convert('RGB')

    # Define a list of possible augmentations
    augmentations = [
        lambda img: TF.rotate(img, random.uniform(-10, 10)),
        lambda img: TF.adjust_brightness(img, random.uniform(0.8, 1.2)),
        lambda img: TF.adjust_contrast(img, random.uniform(0.8, 1.2)),
        lambda img: TF.adjust_saturation(img, random.uniform(0.8, 1.2)),
        lambda img: TF.adjust_hue(img, random.uniform(-0.1, 0.1)),
    ]

    # Randomly choose and apply 2-4 augmentations
    num_augmentations = random.randint(2, 4)
    chosen_augmentations = random.sample(augmentations, num_augmentations)

    for augment in chosen_augmentations:
        image = augment(image)

    # Randomly apply horizontal flip
    if random.random() > 0.5:
        image = TF.hflip(image)

    # Random cropping
    width, height = image.size
    min_dim = min(width, height)
    if min_dim > 256:
        crop_size = random.randint(int(0.8 * min_dim), min_dim)
        top = random.randint(0, height - crop_size)
        left = random.randint(0, width - crop_size)
        image = TF.crop(image, top, left, crop_size, crop_size)

    # Resize to ensure consistent size
    image = TF.resize(image, (224, 224))

    return image

def preprocess_data(examples, processor, augment=False):
    try:
        if isinstance(examples, dict):
            # Batch processing
            image_paths = examples['image_path']
            words = examples['words']
            boxes = examples['bboxes']
            word_labels = examples['ner_tags']
        elif isinstance(examples, list):
            # List of examples
            image_paths = [ex['image_path'] for ex in examples]
            words = [ex['words'] for ex in examples]
            boxes = [ex['bboxes'] for ex in examples]
            word_labels = [ex['ner_tags'] for ex in examples]
        else:
            # Single example
            image_paths = [examples['image_path']]
            words = [examples['words']]
            boxes = [examples['bboxes']]
            word_labels = [ex['ner_tags']]

        print(f"Processing {len(image_paths)} images")

        # Open and convert images
        images = []
        for path in image_paths:
            try:
                img = Image.open(path).convert("RGB")
                images.append(img)
            except Exception as e:
                print(f"Error opening image {path}: {str(e)}")
                raise

        if augment:
            images = [augment_image(img) for img in images]

        # Truncate inputs to max length
        max_length = 512
        truncated_words = [w[:max_length] for w in words]
        truncated_boxes = [b[:max_length] for b in boxes]
        truncated_labels = [l[:max_length] for l in word_labels]

        encoded_inputs = processor(
            images, truncated_words, boxes=truncated_boxes, word_labels=truncated_labels,
            padding="max_length", truncation=True,
            max_length=max_length,
            return_tensors="pt"
        )

        print("Preprocessing completed successfully")
        return encoded_inputs
    except Exception as e:
        print(f"Error in preprocess_data: {str(e)}")
        print(traceback.format_exc())
        raise
```

# Step 3: Model Architecture

```python
# Step 3: Model Architecture
class LayoutLMv3LightningModule(pl.LightningModule):
    def __init__(self, model_name, num_labels, learning_rate):
        super().__init__()
        self.save_hyperparameters()
        self.model = LayoutLMv3ForTokenClassification.from_pretrained(model_name, num_labels=num_labels)
        self.learning_rate = learning_rate
        self.f1_score = F1Score(task="multiclass", num_classes=num_labels, average='macro')

    def forward(self, **inputs):
        return self.model(**inputs)

    def training_step(self, batch, batch_idx):
        outputs = self(**batch)
        loss = outputs.loss
        logits = outputs.logits
        predictions = torch.argmax(logits, dim=2)
        labels = batch['labels']

        mask = labels != -100
        predictions_flat = predictions[mask].flatten()
        labels_flat = labels[mask].flatten()

        f1 = self.f1_score(predictions_flat, labels_flat)
        self.log('train_loss', loss, on_step=True, on_epoch=True, prog_bar=True)
        self.log('train_f1', f1, on_step=True, on_epoch=True, prog_bar=True)
        return loss

    def validation_step(self, batch, batch_idx):
        outputs = self(**batch)
        loss = outputs.loss
        logits = outputs.logits
        predictions = torch.argmax(logits, dim=2)
        labels = batch['labels']

        mask = labels != -100
        predictions_flat = predictions[mask].flatten()
        labels_flat = labels[mask].flatten()

        f1 = self.f1_score(predictions_flat, labels_flat)
        self.log('val_loss', loss, on_step=True, on_epoch=True, prog_bar=True)
        self.log('val_f1', f1, on_step=True, on_epoch=True, prog_bar=True)
        return {"val_loss": loss, "val_f1": f1}

    def configure_optimizers(self):
        optimizer = AdamW(self.parameters(), lr=self.learning_rate)
        scheduler = CosineAnnealingLR(optimizer, T_max=10)
        return {
            "optimizer": optimizer,
            "lr_scheduler": {
                "scheduler": scheduler,
                "interval": "epoch",
                "frequency": 1
            },
        }
```

# Step 4: Training Loop with Cross-Validation

```python
# Step 4: Training Loop with Cross-Validation
def collate_fn(examples, processor):
    return preprocess_data(examples, processor, augment=True)

def collate_fn_val(examples, processor):
    return preprocess_data(examples, processor, augment=False)

def train_model(dataset, processor, num_labels):
    print("Starting model training...")

    try:
        train_dataset = dataset['train'].shuffle(seed=42)
        val_dataset = dataset['test'].shuffle(seed=42)  # Using 'test' as validation since there's no 'validation' split

        print(f"Training dataset size: {len(train_dataset)}")
        print(f"Validation dataset size: {len(val_dataset)}")

        train_loader = DataLoader(
            train_dataset,
            batch_size=4,
            collate_fn=lambda examples: collate_fn(examples, processor),
            num_workers=0,
            pin_memory=True
        )
        val_loader = DataLoader(
            val_dataset,
            batch_size=4,
            collate_fn=lambda examples: collate_fn_val(examples, processor),
            num_workers=0,
            pin_memory=True
        )

        model = LayoutLMv3LightningModule("microsoft/layoutlmv3-base", num_labels=num_labels, learning_rate=2e-5)

        trainer = pl.Trainer(
            max_epochs=20,  # Increase number of epochs
            accelerator='gpu',
            devices=1,
            precision=32,
            logger=True,
            enable_progress_bar=True,
            callbacks=[
                EarlyStopping(monitor="val_loss", patience=5),
                ModelCheckpoint(save_top_k=1, monitor="val_f1", mode="max")
            ],
            accumulate_grad_batches=4  # Gradient accumulation
        )

        trainer.fit(model, train_loader, val_loader)
        return model
    except Exception as e:
        print(f"An error occurred during training:")
        print(str(e))
        print(traceback.format_exc())
        return None
```

# Step 5: Reinforcement Learning Environment

```python
# Step 5: Reinforcement Learning Environment
class InvoiceEnv(gym.Env):
    def __init__(self, dataset, models, processor):
        super(InvoiceEnv, self).__init__()
        self.dataset = dataset["test"]
        self.models = models
        self.processor = processor
        self.current_index = 0
        self.max_index = len(self.dataset) - 1
        self.episode_count = 0
        self.action_space = spaces.Discrete(2)
        self.observation_space = spaces.Box(low=0, high=255, shape=(224, 224, 3), dtype=np.uint8)
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    def reset(self, seed=None, options=None):
        super().reset(seed=seed)
        self.current_index = 0
        self.episode_count += 1
        observation = self._get_observation()
        info = {}
        return observation, info

    def step(self, action):
        if action == 1:
            sample = self.dataset[self.current_index]
            image = Image.open(sample["image_path"]).convert("RGB")
            words = sample["words"]
            boxes = sample["bboxes"]

            encoded_inputs = self.processor(
                image,
                words,
                boxes=boxes,
                return_tensors="pt",
                padding="max_length",
                truncation=True,
                max_length=512
            )
            encoded_inputs = {k: v.to(self.device) for k, v in encoded_inputs.items()}

            predictions = []
            for model in self.models:
                model = model.to(self.device)
                with torch.no_grad():
                    outputs = model(**encoded_inputs)
                predictions.append(outputs.logits.argmax(-1).squeeze().cpu().tolist())

            ensemble_prediction = [max(set(pred), key=pred.count) for pred in zip(*predictions)]

            reward = self.calculate_reward(ensemble_prediction, sample["ner_tags"])
        else:
            reward = 0

        self.current_index += 1
        done = self.current_index > self.max_index
        truncated = False

        if done:
            self.current_index = 0

        observation = self._get_observation()
        info = {}

        return observation, reward, done, truncated, info

    def _get_observation(self):
        sample = self.dataset[self.current_index]
        image = Image.open(sample["image_path"]).convert("RGB")
        image = image.resize((224, 224))
        return np.array(image)

    def calculate_reward(self, predictions, true_labels):
        correct = sum(p == t for p, t in zip(predictions, true_labels))
        total = len(true_labels)
        accuracy = correct / total
        return accuracy
```

# Step 6: RL Training

```python
# Step 6: RL Training
def train_rl_agent(env, total_timesteps=10000):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    env = DummyVecEnv([lambda: env])
    env = VecTransposeImage(env)

    rl_agent = A2C("CnnPolicy", env, verbose=1, tensorboard_log="./a2c_invoice_tensorboard/", device=device)
    rl_agent.learn(total_timesteps=total_timesteps, log_interval=10)
    return rl_agent
```

# Step 7: Evaluation and Visualization

```python
# Step 7: Evaluation and Visualization
def evaluate_model(model, dataset, processor):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    model.eval()

    total_correct = 0
    total_predictions = 0

    for sample in dataset['test']:
        image = Image.open(sample['image_path']).convert("RGB")
        encoded_inputs = processor(image, sample['words'], boxes=sample['bboxes'],
                                   return_tensors="pt", padding="max_length",
                                   truncation=True, max_length=512)
        encoded_inputs = {k: v.to(device) for k, v in encoded_inputs.items()}

        with torch.no_grad():
            outputs = model(**encoded_inputs)

        predictions = outputs.logits.argmax(-1).squeeze().cpu().tolist()
        labels = sample['ner_tags'][:len(predictions)]

        total_correct += sum(p == l for p, l in zip(predictions, labels))
        total_predictions += len(predictions)

    accuracy = total_correct / total_predictions
    print(f"Evaluation Accuracy: {accuracy:.4f}")

    return accuracy

def visualize_predictions(model, processor, image_path):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    image = Image.open(image_path).convert("RGB")
    encoded_inputs = processor(image, return_tensors="pt")
    encoded_inputs = {k: v.to(device) for k, v in encoded_inputs.items()}

    with torch.no_grad():
        outputs = model(**encoded_inputs)

    predictions = outputs.logits.argmax(-1).squeeze().cpu().tolist()
    token_boxes = encoded_inputs['bbox'].squeeze().cpu().tolist()

    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Access the config through the underlying model
    config = model.model.config

    for token_box, prediction in zip(token_boxes, predictions):
        x, y, w, h = token_box
        cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv2.putText(image, config.id2label[prediction], (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36,255,12), 2)

    plt.figure(figsize=(12,16))
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python
def test_custom_image(image_path):
    # Load the saved model and processor
    model = LayoutLMv3ForTokenClassification.from_pretrained("finetuned_LayoutLMv3")
    processor = LayoutLMv3Processor.from_pretrained("finetuned_LayoutLMv3")

    # Prepare the image
    image = Image.open(image_path).convert("RGB")

    # Perform OCR to get words and boxes
    ocr_df = pytesseract.image_to_data(image, output_type='data.frame')
    words = ocr_df['text'].tolist()
    words = [str(w) for w in words if str(w) != 'nan']

    # Get bounding boxes
    boxes = []
    for _, row in ocr_df.iterrows():
        if str(row['text']) != 'nan':
            boxes.append([row['left'], row['top'], row['width'], row['height']])

    # Normalize box coordinates
    width, height = image.size
    normalized_boxes = [[int(1000 * (box[0] / width)),
                         int(1000 * (box[1] / height)),
                         int(1000 * ((box[0] + box[2]) / width)),
                         int(1000 * ((box[1] + box[3]) / height))] for box in boxes]

    # Truncate inputs to max length
    max_length = 512
    words = words[:max_length]
    normalized_boxes = normalized_boxes[:max_length]
    boxes = boxes[:max_length]

    # Encode inputs
    encoding = processor(image, words, boxes=normalized_boxes, truncation=True, max_length=max_length, return_tensors="pt")

    # Try GPU first, fall back to CPU if there's an error
    try:
        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        model.to(device)
        encoding = {k: v.to(device) for k, v in encoding.items()}

        # Perform inference
        with torch.no_grad():
            outputs = model(**encoding)
    except RuntimeError as e:
        print(f"CUDA error encountered: {e}")
        print("Falling back to CPU processing")
        device = torch.device("cpu")
        model.to(device)
        encoding = {k: v.to(device) for k, v in encoding.items()}

        # Perform inference on CPU
        with torch.no_grad():
            outputs = model(**encoding)

    # Get predictions
    predictions = outputs.logits.argmax(-1).squeeze().cpu().tolist()

    # Create a list to store detailed information about each bounding box
    bounding_box_info = []

    # Assuming we have a mapping from label indices to label names
    label_map = {i: f"Label_{i}" for i in range(model.config.num_labels)}

    # Collect information for each bounding box
    for word, box, norm_box, prediction in zip(words, boxes, normalized_boxes, predictions):
        x, y, w, h = box
        bounding_box_info.append({
            "word": word,
            "box_coordinates": {
                "x": x,
                "y": y,
                "width": w,
                "height": h
            },
            "normalized_coordinates": {
                "x1": norm_box[0],
                "y1": norm_box[1],
                "x2": norm_box[2],
                "y2": norm_box[3]
            },
            "predicted_label": label_map[prediction]
        })

    # Visualize predictions
    image_np = np.array(image)

    for info in bounding_box_info:
        box = info["box_coordinates"]
        x, y, w, h = box["x"], box["y"], box["width"], box["height"]
        cv2.rectangle(image_np, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv2.putText(image_np, info["predicted_label"], (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36,255,12), 2)

    plt.figure(figsize=(12,16))
    plt.imshow(image_np)
    plt.axis('off')
    plt.show()

    print("Prediction completed and visualized")

    return bounding_box_info
```
